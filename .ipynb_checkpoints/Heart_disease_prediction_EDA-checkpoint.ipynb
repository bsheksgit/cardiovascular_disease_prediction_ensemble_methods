Let's perform the EDA for the dataset

Univariate - Numerical

```python
def UVA_numeric(data, var_group):
 
    size = len(var_group)
    plt.figure(figsize = (7*size,5*size), dpi = 100)
    p=0
    if size == 4:
        p=1
    else:
        p=2
    #looping for each variable
    for j,i in enumerate(var_group):
    
        # calculating descriptives of variable
        mini = data[i].min()
        maxi = data[i].max()
        ran = data[i].max()-data[i].min()
        mean = data[i].mean()
        median = data[i].median()
        st_dev = data[i].std()
        skew = data[i].skew()
        kurt = data[i].kurtosis()

        # calculating points of standard deviation
        points = mean-st_dev, mean+st_dev

        #Plotting the variable with every information
        plt.subplot(2,2,j+1)
        sns.kdeplot(x=data[i], fill=True)
        sns.lineplot(x=points, y=[0,0], color = 'black', label = "std_dev")
        sns.scatterplot(x=[mini,maxi], y=[0,0], color = 'orange', label = "min/max")
        sns.scatterplot(x=[mean], y=[0], color = 'red', label = "mean")
        sns.scatterplot(x=[median], y=[0], color = 'blue', label = "median")
        plt.xlabel('{}'.format(i), fontsize = 20)
        plt.ylabel('density')
        plt.title('within 1 std_dev = {}; kurtosis = {};\nskew = {}; range = {}\nmean = {}; median = {}'.format((round(points[0],2),round(points[1],2)),
                                                                                                       round(kurt,2),
                                                                                                       round(skew,2),
                                                                                                       (round(mini,2),round(maxi,2),round(ran,2)),
                                                                                                       round(mean,2),
                                                                                                       round(median,2)))
    plt.savefig("UVA_numerical_plot{}".format(p),pad_inches=0.3)
```

```python
UVA_numeric(data,['Height_(cm)', 'Weight_(kg)', 'BMI', 'Alcohol_Consumption'])
```

It can be seen that Weight and BMI have a positive skew and alcohol consumption has many outliers.

```python
UVA_numeric(data,['Fruit_Consumption', 'Green_Vegetables_Consumption', 'FriedPotato_Consumption'])
```

Fruit consumption has many outliers, green_vegetables_consumption and fried_potato_Consumption have a positive skew.

Univariate-Categorical

```python
def UVA_category(data, var_group):

    #setting figure_size
    size = len(var_group)
    plt.figure(figsize = (7*size,7*size), dpi = 100)

    p=0
    if var_group[0] == 'General_Health':
        p=1
    elif var_group[0] == 'Skin_Cancer':
        p=2
    else:
        p=3
        
    # for every variable
    for j,i in enumerate(var_group):
        norm_count = data[i].value_counts(normalize = True)
        n_uni = data[i].nunique()

        #Plotting the variable with every information
        plt.subplot(2,2,j+1)
        sns.barplot(x=norm_count, y=norm_count.index , order = norm_count.index)
        plt.xlabel('fraction/percent', fontsize = 20)
        plt.ylabel('{}'.format(i), fontsize = 20)
        plt.title('n_uniques = {} \n value counts \n {};'.format(n_uni,norm_count))
    plt.savefig("UVA_categorical_plot{}".format(p),pad_inches=0.3)
```

```python
UVA_category(data, ['General_Health', 'Checkup', 'Exercise', 'Heart_Disease'])
```

Most people have very good or good health. Most people have had a health checkup within the last year. Most people exercise. Less than 20% of the people have heart disease. The dataset is not very balanced.

```python
UVA_category(data, ['Skin_Cancer', 'Other_Cancer','Depression', 'Diabetes'])
```

80% of the people don't have skin cancer, other cancer, depression or diabetes.

```python
UVA_category(data, ['Arthritis', 'Sex', 'Age_Category', 'Smoking_History'])
```

Around 30% of the people have diabetes. There are 45% females and 55% males in the data. Most people are in the 65-69 year age category but the distribution of age groups is somewhat uniform.

Bivariate continuous-continuous

```python
numerical_cols = data.select_dtypes(include=['float64'])[:]
numerical_cols.corr()
```

Weight and BMI have a very high correlation. Vegetable and fruit consumption are correlated to some extent.

```python
#taking log of every value to negate outliers
for column in numerical_cols.columns.values:
    mini=1
    if numerical_cols[column].min()<0:
        mini =  abs(numerical_cols[column].min()) + 1
    if column != 'Height_(cm)':
        numerical_cols[column] = [i+mini for i in numerical_cols[column]]
        numerical_cols[column] = numerical_cols[column].map(lambda x : np.log(x))
```

```python
sns.pairplot(numerical_cols)
plt.savefig('Pairplot_numerical', pad_inches=0.3)
```

Only weight and BMI are correlated. Rest don't have as much correlation.

Bivariate Continuous-Categorical

```python
def Bivariate_cont_cat(data,numerical_data,required,serial_num):
    j=1
    for i,col in enumerate(numerical_data.columns.values):
        if i not in required:
            continue
        size = len(numerical_data.columns.values)
        plt.figure(figsize = (10,7), dpi = 100)
        ptable = pd.pivot_table(data=data, values=col, columns='Heart_Disease', aggfunc ='mean')

        plt.subplot(len(required),2,j)
        sns.barplot(ptable)
        plt.ylabel(col)
        plt.xlabel('Heart_disease Yes or No')
        plt.title('Plot of {} vs Heart Disease'.format(col))

        # boxplot
        plt.subplot(len(required),2,j+1)
        sns.boxplot(x='Heart_Disease', y=col, data=data)
        plt.title('Box plot of {}'.format(col))
        j+=2
    plt.savefig('Bivariate Continuous-Categorical-{}'.format(serial_num), pad_inches=0.3)
```

```python
Bivariate_cont_cat(data,numerical_cols,[0,1],1)
```

People with heart diseases have the same mean height but slightly higher mean weight. Height has outliers on both higher and lower side but weight has outliers mainly on the higher side.

```python
Bivariate_cont_cat(data,numerical_cols,[2,3],2)
```

People with heart diseases have a slightly higher mean and median BMI. People with heart diseases have a lower mean alcohol consumption but people without heart diseases have a larger range of alcohol consumption.

```python
Bivariate_cont_cat(data,numerical_cols,[4,5,6],3)
```

Fruit, vegetable and potato consumption is similar

Conclusions:

1. All numerical coloumns except height need to be be converted to log. Bins can be created for height and BMI cols.

2. BMI and weight are highly correlated and hence one of them needs to be dropped.

3. All categorical columns need to be one hot encoded except the general_health column which needs to be label encoded.
